### Splitting the data set

```python
x=data.drop('Species',axis=1)
```

```python
y=data['Species']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
```

### Feature Scaling

```python
from sklearn.preprocessing import StandardScaler  
s = StandardScaler()  
s.fit(x_train)
 
x_train = s.transform(x_train)  
x_test = s.transform(x_test)
```

### Building the model

```python
from sklearn.tree import DecisionTreeClassifier
clf=DecisionTreeClassifier()
clf.fit(x_train,y_train)
```

```python
#Calculating the accuracy of the model
clf.score(x_train,y_train)
```

```python
clf.score(x_test,y_test)
```

```python
y_pred=clf.predict(x_test)
y_pred
```

## We can also use gini criteria

```python
clf_t= DecisionTreeClassifier(criterion = "gini", random_state = 100,max_depth=3, min_samples_leaf=5)
```

```python
clf_t.fit(x_train,y_train)
```

```python
clf_t.score(x_train,y_train)
```

```python
clf_t.score(x_test,y_test)
```


## <span style="font-family: Arial;color:#6B4171;">Predicting the species with SepalLengthCm-4.8,SepalWidthCm-2.8,PetalLengthCm-5.1,PetalWidthCm -1.5 </span> 

```python
#Predicting without using gini criteria
a=clf.predict([[4.8,2.8,5.1,1.5]])
print("The predicted class is {}".format(a[0]))
```

```python
#Predicting with using gini criteria
a1=clf_t.predict([[4.8,2.8,5.1,1.5]])
print("The predicted class is {}".format(a1[0]))
```
